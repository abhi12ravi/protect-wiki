```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
#Get current protection status of article 

def get_protection_status(title):
    
    import requests
    S = requests.Session()

    URL = "https://en.wikipedia.org/w/api.php"

    PARAMS = {
        "action": "query",
        "format": "json",
        "titles": title,
    #     "list": "protectedtitles",
        "prop": "info",
        "inprop": "protection"
    }

    R = S.get(url=URL, params=PARAMS)
    DATA = R.json()

    PAGES = DATA["query"]["pages"]
    
    for k, v in PAGES.items():
        if v['protection'] != []:
            protection_status = v["protection"][0]['level']    
            return protection_status
```

```python
import pandas as pd
df = pd.read_csv('./consolidated-data.csv')

list_protection_statuses = []
  
for index, row in df.iterrows():    
    
    title = row['title']
    protection_status = get_protection_status(title)
    print("Title: ", row['title'], "Protection Status:", protection_status)    
    list_protection_statuses.append(protection_status)

df['Current_Protection_Status'] = list_protection_statuses
```
