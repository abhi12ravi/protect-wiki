```python
def flatten_json(y):
    out = {}

    def flatten(x, name=''):
        if type(x) is dict:
            for a in x:
                flatten(x[a], name + a + '_')
        elif type(x) is list:
            i = 0
            for a in x:
                flatten(a, name + str(i) + '_')
                i += 1
        else:
            out[name[:-1]] = x

    flatten(y)
    return out
```

```python
#Write page id to output file
import csv   
import os, csv
import json



for path, dirs, files in os.walk('.\protectedPageData'):
    for filename in files:
        if (filename.startswith('.')):
            continue
        else:
            ##Add filename to the depp_dataset.csv file
            print ('Processing file...',filename)
            with open (path + "/" + filename,"r") as f:
                data = json.load(f)
            title = data['query']['pages'][filename]['title']
            print(title)
            
            protection_statuses = []
            protection_list_json = data['query']['pages'][filename]['protection']
            for i in range (0,len(protection_list_json)):
                protection_status = protection_list_json[i]['level']
                protection_statuses.append(protection_status)
                
            
            protection_statuses_str = ','.join(protection_statuses)
                
            print('Protection status:', protection_statuses_str)
                       
            filename_list = [filename, title,protection_statuses_str]
            print (filename_list)
            with open(r'hello.csv', 'a', encoding="utf-8") as f:
                writer = csv.writer(f)
                writer.writerow(filename_list)

            
            
```
